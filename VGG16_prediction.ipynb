```python
import cv2
```

```python
image = cv2.imread("dog.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))
image = vgg16.preprocess_input(image)
```

```python
image.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```

```python
import numpy as np
```

```python
image = np.expand_dims(image, 0)
```

```python
image.shape
```

```python
vgg_result = vgg_model.predict(image)
```

```python
vgg16.decode_predictions(vgg_result, top=5)
```
