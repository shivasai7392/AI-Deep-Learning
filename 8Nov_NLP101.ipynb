```
import string
```

```
punc_table = str.maketrans("","",string.punctuation)
```

```
type(punc_table)
```

```
cleaned_txt = txt.translate(punc_table).replace("\n","")
```

```
print(cleaned_txt)
```

```
cleaned_words = word_tokenize(cleaned_txt)
```

```
cleaned_words
```

```
len(cleaned_words)
```

```
from nltk.corpus import stopwords
```

```
 nltk.download('stopwords')
```

```
stopwords.words("english")
```

```
doubled_cleaned_words = [word for word in cleaned_words if word.lower() not in stopwords.words("english")]
```

```
len(doubled_cleaned_words)
```

```
from nltk import FreqDist
```

```
fd = FreqDist(doubled_cleaned_words)
```

```
fd
```

```
fd.most_common(10)
```

```
len(fd.most_common())
```

```
import spacy
```

```
spacytext = "For example, if the option seller is explicitly writing calls on Tesla the contract will state that if “Tesla’s shares reach a stock price higher than X by a specific date, then you can buy the Tesla shares off of me for a strike price of X”."
```

```
print(spacytext)
```

```
nlp = spacy.load("en")
doc = nlp(spacytext)
```

```
type(doc)
```

```
for sentence in doc.sents:
  print(sentence)
```

```
for word in doc:
  print(word)
```

```
for word in doc:
  print(word,word.is_punct,word.is_stop)
```

```
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```
doc.ents
```

```
from spacy import displacy
```

```
displacy.render(doc,style="ent",jupyter=True)
```

```
displacy.render(doc,style="dep",jupyter=True)
```
