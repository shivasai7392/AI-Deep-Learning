```python
import numpy as np 
```

```python
x1 = 3
x2 = 1

y = 16
```

```python
w1 = np.random.randint(1, 10, 1)
w2 = np.random.randint(1, 10, 1)
print(w1, w2)
y_pred = x1*w1 + x2*w2  
```

```python
y_pred 
```

```python
error = (y_pred - y)**2
dEdw1 = 2*(y_pred - y)*(x1)
dEdw2 = 2*(y_pred - y)*(x2)
```

```python
w1 = w1 -dEdw1
w2 = w2 -dEdw2
```

```python
w1, w2
```

```python
# With one iteration they wont work - you have to repeat this process ... for loop 
```

```python
x1 = 3
x2 = 1

y = 16
```

```python
w1 = np.random.randint(1, 10, 1)
w2 = np.random.randint(1, 10, 1)
print(w1, w2)
alpha = 0.01

for epoch in range(100):
    y_pred = x1*w1 + x2*w2
    E = (y_pred - y)**2
    dEdw1 = 2*(y_pred - y)*x1
    dEdw2 = 2*(y_pred - y)*x2
    
    w1 = w1 - alpha*dEdw1
    w2 = w2 - alpha*dEdw2
    
    print(f"y_pred = {y_pred[0]} and y is {y}")
#     print(f"w1 = {w1} and w2 = {w2}")
```
